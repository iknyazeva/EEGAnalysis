## Анализ воспроизводимости эффектов, обнаруживаемых в результате ЭЭГ исследования

Исследовательский вопрос: 
 - Когда я получаю в результате анализа какой-то результат, имея на руках информацию о размере выборки, и статистиках эффекта чему и насколько можно верить? 
 - Можно ли получить какие-то цифры для оценки неопределенности?


Как предлагается это изучать?
 1. Метод оценок распределения воспроизводства уровней значимости. Предлагается выбрать подмножество испытуемых, оценить для них разницу между состояниями, упорядочить эти разницы по увеличению p-value. Затем для разных подвыборок такого же размера посмотреть какие p-value  получаются, будет ли наблюдаться следующая тенденция: чем меньше p-value, тем более достоверна полученная разница между состояниями
 
2. Воспроизводство "значимых" эффектов. Для набора подвыборок оценить частоту воспроизводства эффекта. Посмотреть, есть ли процесс "усиления" рисунка с ростом выборки. В этом случае, можно например осторожно аргументировать в пользу повышения допустимого порога при небольших выборках (альфа уровень)
3. Посмотреть байесовский доверительный интервал и его динамику при росте количества испытуемых.
4. Перейти от анализа парных взаимосвязей к паттернам, смотреть не отдельно воспроизводство конкретного ребра, а паттерна целиком

```python
df = pd.read_csv('eeg_dataframe_nansfilled.csv', index_col=0)
analyzer = EEGPairedPermutationAnalyser(data_df=df, num_perm=1000, thres=0.001)
```

```python
top_k=40
num_exps=100
num_perms=10000
```

```python
analyzer.key_bands[7]
```

```python
size=30
for band in tqdm(range(1,8)):
    sign_df = analyzer.p_val_reproducibility(size=size, band=band,
                                            num_perms=5000, num_exps=num_exps, is_param=True)
    sorted_df = sign_df.sort_values(by=['Orig','Q_50'], ascending=False)
    plot_df = sorted_df.head(top_k).reset_index().melt(id_vars='index')
    plot_df.columns = ['ChanPair', 'Type', 'Significance']
    fig,ax = plt.subplots(figsize= (12,4))
    fig.autofmt_xdate(rotation=45)
    sns.boxplot(x="ChanPair", y="Significance", data=plot_df, ax=ax)
    ax.set_title(f"Singificance reproducibility for {analyzer.key_bands[band]} band, group size {size}, top {top_k} most significant\n \
    paired t-test stat, {num_exps} subgroups");
```

```python
for band in tqdm(range(1,8)):
    sign_df = analyzer.p_val_reproducibility(size=size, band=band,
                                            num_perms=num_perms, num_exps=num_exps, is_param=False)
    sorted_df = sign_df.sort_values(by=['Orig','Q_50'], ascending=False)
    plot_df = sorted_df.head(top_k).reset_index().melt(id_vars='index')
    plot_df.columns = ['ChanPair', 'Type', 'Significance']
    fig,ax = plt.subplots(figsize= (12,4))
    fig.autofmt_xdate(rotation=45)
    sns.boxplot(x="ChanPair", y="Significance", data=plot_df, ax=ax)
    ax.set_title(f"Singificance reproducibility for {analyzer.key_bands[band]} band, group size {size}, top {top_k} most significant\n \
    permutation stat with number of permutations {num_perms} ");
```
